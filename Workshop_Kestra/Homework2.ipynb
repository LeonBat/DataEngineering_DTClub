**Question 3**

How many rows are there for Yellow Taxi data for all CSV files in year 2020?

To answer that question I ran the pipeline for the important date and queried the database



````python
SELECT COUNT(*) 
 FROM yellow_tripdata 
 WHERE tpep_pickup_datetime >= '2020-01-01' 
 AND tpep_pickup_datetime < '2021-01-01';
````

The result indicated 24,648,299 rows

**Question 5**


How many rows are there for the Yellow Taxi data for the March 2021 CSV file?


````python
SELECT COUNT(*) AS row_count
 FROM public.yellow_tripdata
 WHERE tpep_pickup_datetime >= TIMESTAMP '2021-03-01'
AND tpep_pickup_datetime <  TIMESTAMP '2021-04-01';
````

The result indicated 1,925,152 rows
